```python
def template_relative_syllogistic(template_name):
  templates = {
      "noun_noun_si": "{} {} who is a {} is a {}",
      "noun_adj_si": "{} {} who is a {} is {}",
      "noun_noun_pl": "{} {} who are {} are {}",
      "noun_adj_pl": "{} {} who are {} are {}",
      "adj_noun_si": "{} {} who is {} is a {}",
      "adj_adj_si": "{} {} who is {} is {}",
      "adj_noun_pl": "{} {} who are {} are {}",
      "adj_adj_pl": "{} {} who are {} are {}",
      "noun_neg_noun_si": "{} {} who is not a {} is a {}",
      "noun_neg_adj_si": "{} {} who is not a {} is {}",
      "noun_neg_noun_pl": "{} {} who are not {} are {}",
      "noun_neg_adj_pl": "{} {} who are not {} are {}",
      "adj_neg_noun_si": "{} {} who is not {} is a {}",
      "adj_neg_adj_si": "{} {} who is not {} is {}",
      "adj_neg_noun_pl": "{} {} who are not {} are {}",
      "adj_neg_adj_pl": "{} {} who are not {} are {}",
      "noun_noun_neg_si": "{} {} who is a {} is not a {}",
      "noun_adj_neg_si": "{} {} who is a {} is not {}",
      "noun_noun_neg_pl": "{} {} who are {} are not {}",
      "noun_adj_neg_pl": "{} {} who are {} are not {}",
      "adj_noun_neg_si": "{} {} who is {} is not a {}",
      "adj_adj_neg_si": "{} {} who is {} is not {}",
      "adj_noun_neg_pl": "{} {} who are {} are not {}",
      "adj_adj_neg_pl": "{} {} who are {} are not {}",
      "noun_neg_noun_neg_si": "{} {} who is not a {} is not a {}",
      "noun_neg_adj_neg_si": "{} {} who is not a {} is not {}",
      "noun_neg_noun_neg_pl": "{} {} who are not {} are not {}",
      "noun_neg_adj_neg_pl": "{} {} who are not {} are not {}",
      "adj_neg_noun_neg_si": "{} {} who is not {} is not a {}",
      "adj_neg_adj_neg_si": "{} {} who is not {} is not {}",
      "adj_neg_noun_neg_pl": "{} {} who are not {} are not {}",
      "adj_neg_adj_neg_pl": "{} {} who are not {} are not {}",
      "neg_noun_noun_si": "{} non-{} who is a {} is a {}",
      "neg_noun_adj_si": "{} non-{} who is a {} is {}",
      "neg_noun_noun_pl": "{} non-{} who are {} are {}",
      "neg_noun_adj_pl": "{} non-{} who are {} are {}",
      "neg_adj_noun_si": "{} non-{} who is {} is a {}",
      "neg_adj_adj_si": "{} non-{} who is {} is {}",
      "neg_adj_noun_pl": "{} non-{} who are {} are {}",
      "neg_adj_adj_pl": "{} non-{} who are {} are {}",
      "neg_noun_neg_noun_si": "{} non-{} who is not a {} is a {}",
      "neg_noun_neg_adj_si": "{} non-{} who is not a {} is {}",
      "neg_noun_neg_noun_pl": "{} non-{} who are not {} are {}",
      "neg_noun_neg_adj_pl": "{} non-{} who are not {} are {}",
      "neg_adj_neg_noun_si": "{} non-{} who is not {} is a {}",
      "neg_adj_neg_adj_si": "{} non-{} who is not {} is {}",
      "neg_adj_neg_noun_pl": "{} non-{} who are not {} are {}",
      "neg_adj_neg_adj_pl": "{} non-{} who are not {} are {}",
      "neg_noun_noun_neg_si": "{} non-{} who is a {} is not a {}",
      "neg_noun_adj_neg_si": "{} non-{} who is a {} is not {}",
      "neg_noun_noun_neg_pl": "{} non-{} who are {} are not {}",
      "neg_noun_adj_neg_pl": "{} non-{} who are {} are not {}",
      "neg_adj_noun_neg_si": "{} non-{} who is {} is not a {}",
      "neg_adj_adj_neg_si": "{} non-{} who is {} is not {}",
      "neg_adj_noun_neg_pl": "{} non-{} who are {} are not {}",
      "neg_adj_adj_neg_pl": "{} non-{} who are {} are not {}",
      "neg_noun_neg_noun_neg_si": "{} non-{} who is not a {} is not a {}",
      "neg_noun_neg_adj_neg_si": "{} non-{} who is not a {} is not {}",
      "neg_noun_neg_noun_neg_pl": "{} non-{} who are not {} are not {}",
      "neg_noun_neg_adj_neg_pl": "{} non-{} who are not {} are not {}",
      "neg_adj_neg_noun_neg_si": "{} non-{} who is not {} is not a {}",
      "neg_adj_neg_adj_neg_si": "{} non-{} who is not {} is not {}",
      "neg_adj_neg_noun_neg_pl": "{} non-{} who are not {} are not {}",
      "neg_adj_neg_adj_neg_pl": "{} non-{} who are not {} are not {}",
  }

  return templates[template_name]

def natural_language_sentence_generation(type, quantifier, variables, negations):
  det = None
  if quantifier == "all":
    det = random.choice(["all", "every"])
  elif quantifier == "exists":
    det = random.choice(["some", "a"])

  if det:
    negs = [negations[variables[0]], negations[variables[1]],negations[variables[2]]]
    sing = "pl" if det in ["all", "some"] else "si"

    template_id = ""
    if negations[variables[0]] == True :
      template_id += "neg_"

    if type in ["NA", "NN"]:
      template_id += "noun_"
    else :
      template_id += "adj_"

    if negations[variables[1]] == True :
      template_id += "neg_"
    
    if type in ["NN", "AN"]:
      template_id += "noun_"
    else :
      template_id += "adj_"

    if negations[variables[2]] == True :
      if quantifier == "all":
        det = "no"
      else :
        template_id += "neg_"

    if sing == "pl":
      template_id += "pl"
    else :
      template_id += "si"

  return template_relative_syllogistic(template_id).format(det, variables[0], variables[1], variables[2])

    


def relative_syllogistic_generate_logic_str(quantifier, variables, negations):
  
  variables = ["-"+x if negations[x] else x for x in variables]
  if quantifier == "all":
    return "all x.(({}(x) & {}(x)) -> {}(x))".format(variables[0], variables[1], variables[2])
  elif quantifier == "exists":
    return "exists x.({}(x) & {}(x) & {}(x))".format(variables[0], variables[1], variables[2])
  


def post_process(sentence, random_type):
  if sentence.split()[0] in ["every", "all"]:
    if (sentence.split()[-2] == "not") or (sentence.split()[-3] == "not"):
      sentence = sentence.replace("every", "no")
      sentence = sentence.replace("all", "no")
      if random_type[-1] == "N":
        sentence = sentence.replace("are", "is a")
      else :
        sentence = sentence.replace("are", "is")

      sent_list = sentence.split()
      if sent_list[-2] == "not":
        del sent_list[-2]
      if sent_list[-3] == "not":
        del sent_list[-3]

      sentence = " ".join(sent_list)

  return sentence


def generate_sentence_relative_syllogistic(nouns, adjs, domain):
  template_types = ["NN", "NA", "AN", "AA"]

  random_type = random.choice(template_types)

  if random_type == "NN":
    variables = random.sample(nouns, 3)
  elif random_type == "NA":
    variables = random.sample(nouns, 2)
    variables.append(random.choice(adjs))
  elif random_type == "AN":
    variables = [random.choice(nouns), random.choice(adjs), random.choice(nouns)]
    while (variables[0] == variables[2]):
      variables = [random.choice(nouns), random.choice(adjs), random.choice(nouns)]
  else :
    variables = [random.choice(nouns)]
    variables.extend(random.sample(adjs, 2))

  negations = {variables[0] : random.choice([True, False]), 
               variables[1] : random.choice([True, False]),
               variables[2] : random.choice([True, False])}

  
  random_quantifier = random.choice(["all", "exists"])

  
  logic = relative_syllogistic_generate_logic_str(random_quantifier, variables, negations)
  sentence = natural_language_sentence_generation(random_type, random_quantifier, variables, negations)
  
  return logic, sentence, random_quantifier


def model_generation(domain, signature, prob = 0.5):
  model = [(d, d) for d in domain]
  model.append(('people', set(domain)))
  for f in signature :
    sample = []
    for v in domain:
      if random.uniform(0, 1) <= prob :
        sample.append(v)
    model.append((f, set(sample)))

  return model

def model_to_nl_without_verbs(model, relations, domain):

  str_temp = ""

  for i in model:
    func = i[0]
    values = list(i[1])
    if func in domain :
      continue
    if func == "people":
        values_str = ""
        for str_val in values[:-1]:
          values_str += str_val + ", "
        
        str_temp += "{}and {} are people in the group. ".format(values_str, values[-1])

    if func in relations:
      
      if len(values) == 0:
        str_temp += "There are no {}. ".format(func)
      elif len(values) == 1:
        str_temp += "{} is a {}. ".format(values[0], func)
      else :
        values_str = ""
        for str_val in values[:-1]:
          values_str += str_val + ", "
        str_temp += "{}and {} are {}. ".format(values_str, values[-1], func)

  str_model = str_temp.replace(", and ", " and ")
  return str_model

def generate_datapoint(nouns, adjectives, names, domain_size = 5, signature_size = 4, max_count = 100, label = "random", connectors = 3):

  domain = random.sample(names, random.randrange(3, domain_size))
  #domain = random.sample(names, domain_size)
  nouns = random.sample(nouns, random.randrange(3,signature_size))
  adjs = random.sample(adjectives, random.randrange(2,signature_size))

  #nouns = random.sample(nouns, signature_size + 2)
  #adjs = random.sample(adjectives, signature_size + 2)


  boolean_connector = random.choice(['AND', 'OR'])
  logic_set = []
  sentence_set = []
  quantifier_set = []

  for i in range(connectors):
    logic, sentence, quantifiers = generate_sentence_relative_syllogistic(nouns, adjs, domain)
    logic_set.append(logic)
    sentence_set.append(sentence)
    quantifier_set.append(" ".join(quantifiers))


  if boolean_connector == "AND": 
    logic = "(" + ") & (".join(logic_set) + ")"
    sentence = " and ".join(sentence_set)
  else :
   logic = "(" + ") | (".join(logic_set) + ")" 
   sentence = " or ".join(sentence_set)

  signature = nouns.copy()
  signature.extend(adjs)

  model_found = False
  model = None
  model_nl = None

  if label == "random":
    label = random.choice([True, False])
  
  if quantifier != "assignment":
    counter = 0
    while(True):
      model = model_generation(domain, signature)

      val = nltk.Valuation(model)
      dom = val.domain
      m = nltk.Model(dom, val)
      g = nltk.sem.Assignment(dom)

      validity = m.evaluate(logic, g)

      if validity == label:
        model_nl = model_to_nl_without_verbs(model, signature, domain)
        model_found = True
        break
      elif counter > max_count:
        break
      counter += 1

    return logic, sentence, model, label, model_found, model_nl, boolean_connector, quantifier_set

def generate_data(nouns, adjectives, names, domain_size = 5, signature_size = 5, max_count = 100, num_data_points = 100000):
  data = {
      'model' : [],
      'model_nl' : [],
      'sentence' : [],
      'logic' : [],
      'label' : [],
      'boolean_connector': [],
      'quantifiers' : []
  }

  count = 0
  model_not_found = 0
  count_none = 0

  count_false = 0
  count_truth = 0
  label = "random"
  while(count < num_data_points):
    logic, sentence, model, label, model_found, model_nl, boolean_connector, quantifiers = generate_datapoint(nouns, adjectives, names, domain_size, signature_size, max_count, label)
    if (model_nl == None) or ("There are no" in model_nl) or ("No one" in model_nl):
      if model_nl == None :
        count_none += 1
      model_not_found += 1
      continue
    else:
      if model_found :
        data['model'].append(model)
        data['sentence'].append(sentence)
        data['logic'].append(logic)
        data['label'].append(label)
        data['model_nl'].append(model_nl)
        data['boolean_connector'].append(boolean_connector)
        data['quantifiers'].append(",".join(quantifiers))
        count += 1
        if label == False:
          count_false += 1
        else :
          count_truth += 1
        
        if count_false >= num_data_points/2:
          label = True
        elif count_truth >= num_data_points/2:
          label = False
        else :
          label = "random"
      else :
        model_not_found += 1
    if (count % 20000) == 0 :
      print("{} data points are generated".format(count))
  
  print("Number of times model did not generated: {}".format(model_not_found))
  print("None model count: {}".format(count_none))

  df = pd.DataFrame(data)
  df['model'] = df['model'].astype(str)
  df = df.drop_duplicates(subset = ['sentence', 'model'], keep = 'first')

  return df
```
